## Основные концепции autograd

Автоматическое дифференцирование (autograd) в PyTorch является одной из ключевых возможностей, которая делает PyTorch мощным инструментом для разработки и обучения нейронных сетей. Далее мы рассмотрим, как работает autograd в PyTorch, объясним основные концепции с примерами и формулами.

`autograd` — это система автоматического дифференцирования в PyTorch, которая позволяет автоматически вычислять градиенты тензоров, что необходимо для оптимизации моделей машинного обучения. Она строит вычислительный граф на основе операций, выполняемых с тензорами, и использует его для вычисления градиентов при обратном проходе (backward pass).

__Основные концепции autograd__

- Тензоры с поддержкой градиентов: В PyTorch градиенты хранятся в атрибуте `.grad` тензора. Чтобы тензор участвовал в вычислении градиентов, при его создании необходимо установить `requires_grad=True`.

- Вычислительный граф: Каждая операция с тензорами создает узлы в вычислительном графе. Граф строится динамически во время выполнения (динамический граф вычислений).

- Обратный проход: При вызове метода `.backward()` PyTorch автоматически вычисляет градиенты всех тензоров, участвовавших в вычислении целевой функции.

### Пример 1. Простая функция

````python
# Создаем тензор с поддержкой градиентов
x = torch.tensor(2.0, requires_grad=True)

# Определяем функцию y = x^2
y = x ** 2
````

````python
print(x.grad)
````

````python
# Выполняем обратный проход для вычисления градиента dy/dx
y.backward()

# Выводим градиент
print(f'Градиент dy/dx при x=2: {x.grad}')  # Вывод: 4.0
````

__Объяснение:__

   1. Вычисление прямого прохода:
   $$ 
   y = x^2 = (2)^2 = 4 
   $$

   2. Вычисление градиента:
   $$
   \frac{dy}{dx} = 2x = 4
   $$

После вызова `y.backward()`, градиент `dy/dx` сохраняется в `x.grad`.



````python
# Визуализация вычислительного графа
import torch
from torchviz import make_dot
# pip install torchviz
# sudo apt install graphviz

# Создаем визуализацию графа
dot = make_dot(y, params={'x': x, 'y': y})
# dot.render("computational_graph_2", format="png")

dot
````

### Пример 2. Сложная функция

````python
# Рассмотрим более сложный пример с несколькими операциями.

# Создаем тензоры с поддержкой градиентов
x = torch.tensor(3.0, requires_grad=True)
w = torch.tensor(4.0, requires_grad=True)
b = torch.tensor(5.0, requires_grad=True)

# Определяем функцию y = w * x + b
y = w * x + b

# Выполняем обратный проход для вычисления градиентов
y.backward()

# Выводим градиенты
print(f'Градиент dy/dx: {x.grad}')
print(f'Градиент dy/dw: {w.grad}')
print(f'Градиент dy/db: {b.grad}')
````

$$

\frac{dy}{dw} = x = 3
$$
$$
\frac{dy}{db} = 1
$$

После вызова `y.backward()`, градиенты сохраняются в соответствующих тензорах `x.grad`, `w.grad` и `b.grad`.

````python
# Использование цепного правила

# Создаем тензоры с поддержкой градиентов
a = torch.tensor(2.0, requires_grad=True)
b = torch.tensor(3.0, requires_grad=True)

# Определяем сложную функцию: y = (a + b) * (a * b)
y = (a + b) * (a * b)

# Выполняем обратный проход
y.backward()

# Выводим градиенты
print(f'Градиент dy/da: {a.grad}')  # Вывод: 3*(a + b) + (a * b)*1 = 3*(2+3) + (2*3)*1 = 15 + 6 = 21
print(f'Градиент dy/db: {b.grad}')  # Аналогично
````

### Формулы и математическое обоснование

Рассмотрим, как `autograd` использует цепное правило для вычисления градиентов.

__Цепное правило__

Если функция $ y $ зависит от $ u $, а $ u $ зависит от $ x $, то:

$$
\frac{dy}{dx} = \frac{dy}{du} \cdot \frac{du}{dx}
$$

Autograd в PyTorch применяет это правило рекурсивно к каждому узлу вычислительного графа.



#### Пример с цепным правилом

Рассмотрим функцию:

$$
y = (a + b) \cdot (a \cdot b)
$$

Распишем поэтапно:

1. $ u = a + b $
2. $ v = a \cdot b $
3. $ y = u \cdot v $

Вычислим градиенты:

$$
\frac{dy}{du} = v
$$
$$
\frac{dy}{dv} = u
$$
$$
\frac{du}{da} = 1
$$
$$
\frac{du}{db} = 1
$$
$$
\frac{dv}{da} = b
$$
$$
\frac{dv}{db} = a
$$

Теперь применим цепное правило:

$$
\frac{dy}{da} = \frac{dy}{du} \cdot \frac{du}{da} + \frac{dy}{dv} \cdot \frac{dv}{da} = v \cdot 1 + u \cdot b
$$
$$
\frac{dy}{db} = \frac{dy}{du} \cdot \frac{du}{db} + \frac{dy}{dv} \cdot \frac{dv}{db} = v \cdot 1 + u \cdot a
$$

Подставим значения $ a = 2 $, $ b = 3 $:

$$
u = 2 + 3 = 5
$$
$$
v = 2 \cdot 3 = 6
$$
$$
\frac{dy}{da} = 6 \cdot 1 + 5 \cdot 3 = 6 + 15 = 21
$$
$$
\frac{dy}{db} = 6 \cdot 1 + 5 \cdot 2 = 6 + 10 = 16
$$

Эти значения соответствуют результатам из примера выше.

### Важные аспекты autograd



1. Динамический граф вычислений

    В отличие от некоторых других фреймворков, PyTorch использует динамический граф вычислений, что означает, что граф строится "на лету" во время выполнения программы. Это позволяет гибко изменять структуру модели во время обучения.

2. Управление памятью

    PyTorch автоматически освобождает память, занятую промежуточными результатами вычислений, после выполнения обратного прохода. Это предотвращает утечки памяти и делает использование памяти более эффективным.

3. Отключение `autograd`

    В некоторых случаях, например, при инференсе (предсказании без обучения), вычисление градиентов не требуется. В таких случаях можно отключить autograd для экономии ресурсов:

    ```python
    with torch.no_grad():
        y_pred = model(x)
    ```

4. Zeroing градиентов

    Перед выполнением обратного прохода необходимо обнулить градиенты, иначе они будут суммироваться:

    ```python
    optimizer.zero_grad()
    ```

````python
import torch

# Создаем тензор с requires_grad=True
x = torch.tensor([2.0, 3.0], requires_grad=True)

# Выполняем операцию
y = x ** 2  # y = [4.0, 9.0]

# Проверяем grad_fn для y
print(y.grad_fn)  # <PowBackward0 object at ...>

# Выполняем обратное распространение
y.sum().backward(retain_graph=True)

# Проверяем grad для x
print(x.grad)  # tensor([4.0, 6.0])

# Проверяем grad_fn для x
print(x.grad_fn)  # None, так как x - начальный тензор
````
